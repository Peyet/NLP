```python
for i in range(161,254):
    for j in range(161,254):
        s = hex(i) + hex(j)
        s = s.replace('0x','')
        bi = [int(s[0:2],16),]
        bi = bi + [int(s[2:4],16),]
        bs = bytearray(bi)
        print(bytes(bs).decode('gb18030'),i,j,'  ',end='')
        if j==9:
            print('\n')
```
